## Matplotlib

To jedna z bibliotek do wizualizacji.

```python
import matplotlib.pyplot as plt
# specjalna komenda, która powoduje wklejanie do notebooka
%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0, 10, 100) # ndarray
y = np.sin(x)
x, y
```

```python
plt.plot(x, y);
```

Wykres składa się z kilku elementów:
* `figure` - obiekt z osiami, może zawierać wiele rysunków
* `axis` - osie danych w których dana krzywa jest osadzona
* `line` - linia na wykresie; w osiach mogą znajdować się też inne rzeczy jak np. punkty

W notebooku `figure` i `axis` tworzone są dla jednego rysunku automatycznie, więc jeden rysunek definiowany jest w jednej komórce.

```python
y2 = np.cos(x) # kolejna seria danych
y3 = np.cos(x)+1
plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x, y3);
plt.savefig('pic.jpg')
```

Serie danych tworzą się automatycznie z kolejnymi kolorami. Możemy zdefiniować styl linii używając trzeciego argumentu w postaci tekstu; więcej informacji w [dokumentacji](http://matplotlib.org/1.5.3/api/pyplot_api.html#matplotlib.pyplot.plot).

```python
plt.plot(x, y, 'gh')
plt.plot(x, y2, 'gH')
plt.plot(x, y3, 'bH');
```

```python
plt.plot(x, y, 'r')
plt.plot(x, y2, 'H')
plt.title('Sinus i cosinus')
plt.legend(['sin(x)', 'cos(x)']) 
plt.xlabel('x')
plt.ylabel('y');
```

Możliwe jest tworzenie podwykresów.

```python
plt.subplot(221)
plt.plot(x, y, 'r:')
plt.legend(['a'])
plt.ylabel('y1')

#plt.subplot(122)
#plt.plot(x, y2, 'c+');
#plt.legend(['a'])
#plt.xlabel('x')
#plt.ylabel('y2');
```
