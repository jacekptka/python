## Matplotlib

To jedna z bibliotek do wizualizacji.

```python
import matplotlib.pyplot as plt
# specjalna komenda, kt√≥ra powoduje wklejanie do notebooka
%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0, 10, 100) # ndarray
y = np.sin(x)
x, y
```

```python
plt.plot(x, y);
```

Wykres sk≈Çada siƒô z kilku element√≥w:
* `figure` - obiekt z osiami, mo≈ºe zawieraƒá wiele rysunk√≥w
* `axis` - osie danych w kt√≥rych dana krzywa jest osadzona
* `line` - linia na wykresie; w osiach mogƒÖ znajdowaƒá siƒô te≈º inne rzeczy jak np. punkty

W notebooku `figure` i `axis` tworzone sƒÖ dla jednego rysunku automatycznie, wiƒôc jeden rysunek definiowany jest w jednej kom√≥rce.

```python
y2 = np.cos(x) # kolejna seria danych
y3 = np.cos(x)+1
plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x, y3);
plt.savefig('pic.jpg')
```

Serie danych tworzƒÖ siƒô automatycznie z kolejnymi kolorami. Mo≈ºemy zdefiniowaƒá styl linii u≈ºywajƒÖc trzeciego argumentu w postaci tekstu; wiƒôcej informacji w [dokumentacji](http://matplotlib.org/1.5.3/api/pyplot_api.html#matplotlib.pyplot.plot).

```python
plt.plot(x, y, 'gh')
plt.plot(x, y2, 'gH')
plt.plot(x, y3, 'bH');
```

```python
plt.plot(x, y, 'r')
plt.plot(x, y2, 'H')
plt.title('Sinus i cosinus')
plt.legend(['sin(x)', 'cos(x)']) 
plt.xlabel('x')
plt.ylabel('y');
```

Mo≈ºliwe jest tworzenie podwykres√≥w.

```python
plt.subplot(221)
plt.plot(x, y, 'r:')
plt.legend(['a'])
plt.ylabel('y1')

#plt.subplot(122)
#plt.plot(x, y2, 'c+');
#plt.legend(['a'])
#plt.xlabel('x')
#plt.ylabel('y2');
```
