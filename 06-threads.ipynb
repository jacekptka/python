```python
import _thread
import time

def thread_func(threadName, delay):
   count = 0
   while count < 5:
      time.sleep(delay)
      count += 1
      print(threadName + "  " + time.ctime(time.time()))
    
thread_func("1", 1)
thread_func("2", 2)
print("W" + "  " + time.ctime(time.time()))
```

```python
t1=_thread.start_new_thread(thread_func, ("1", 1))
t2=_thread.start_new_thread(thread_func, ("2", 2))
print("W" + "  " + time.ctime(time.time()))
```
